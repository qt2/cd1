```python
import matplotlib.pyplot as plt
import numpy as np

# 
labels = ["100", "1000", "10000", "100000"]
values = {
    "List": [43, 45, 58, 46],
    "FastList": [567, 421, 427, 1566],
    "Matrix": [46, 41, 46, 56],
}

x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots()

# 
bars1 = ax.bar(x - width, values["List"], width, label="List")
bars2 = ax.bar(x, values["FastList"], width, label="FastList")
bars3 = ax.bar(x + width, values["Matrix"], width, label="Matrix")

# 
ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)
ax.bar_label(bars3, padding=3)

# 
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Calculation time (s)")
ax.set_title("Calculation time of checking the existence of 100,000 edges")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.tight_layout()
plt.savefig("part01_time.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
labels = ["100", "1000", "10000", "100000"]
values = {
    "List": [17,179,1796,17968],
    "FastList": [16,185,1873,18743],
    "Matrix": [9,976,97656,9765625],
}

x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots()

# 
bars1 = ax.bar(x - width, values["List"], width, label="List")
bars2 = ax.bar(x, values["FastList"], width, label="FastList")
bars3 = ax.bar(x + width, values["Matrix"], width, label="Matrix")

# 
ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)
ax.bar_label(bars3, padding=3)

# 
ax.set_xlabel("Number of vertices")
ax.set_ylabel("Memory usage (KB)")
ax.set_title("Memory usage of each data structure")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 100000)
ax.legend()

plt.tight_layout()
plt.savefig("part01_memory.png", dpi=300)
plt.show()
```
